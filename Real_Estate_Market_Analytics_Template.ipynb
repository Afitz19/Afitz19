```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('https://econdata.s3-us-west-2.amazonaws.com/Reports/Hotness/RDC_Inventory_Hotness_Metrics_County_History.csv', nrows=5000)
```

```python
df.astype({'county_name':'object','cbsa_title': 'object','hotness_score':'float','supply_score':'float','demand_score':'float','median_dom_vs_us':'float'})
```

```python
sellers_market = (df.loc[:, 'hotness_score'] < 75) & (df.loc[:,'supply_score'] < 75) & (df.loc[:,'demand_score'] < 75)
df1m = df[sellers_market]
```

```python
df1m.plot.scatter(x='hotness_rank', y='median_days_on_market')
```

```python
df1m.plot.scatter(x='hotness_score', y='demand_score')
```

```python
df1m.plot.scatter(x='hotness_score', y='hotness_rank')
```
