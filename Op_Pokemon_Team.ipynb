```python
import pandas as pd 
from pandas import DataFrame
```

```python
df = pd.read_csv('/content/pokemon dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
op_team_draft = df.loc[0:,['name','attack','base_total','defense','experience_growth','hp','sp_attack','sp_defense','speed','type1','type2','generation']]
```

```python
op_team_draft
```

```python
op_team = op_team_draft.fillna('No second type')
```

```python
op_team
```

```python
Elite_Mostly_legendary_pokemon = op_team.loc[(op_team.attack > 95) & (op_team.base_total > 100) & (op_team.hp > 90) & (op_team.sp_attack > 100) & (op_team.sp_defense > 50) & (op_team.speed > 50) & (op_team.defense > 50)]
```

```python
Strong_Pokemon = op_team.loc[(op_team.attack > 95) & (op_team.base_total > 100) & (op_team.hp < 100) & (op_team.sp_attack > 85) & (op_team.sp_defense > 50) & (op_team.speed > 50) & (op_team.defense > 50)]
```

```python
Target_pokemon_teams_for_pokemon_games = Elite_Mostly_legendary_pokemon.append(Strong_Pokemon)
Target_pokemon_teams_for_pokemon_games
```
